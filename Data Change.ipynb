# Transform Dataframes

## Fat Supply Data

```python
# Create a filtered dataframe from specific columns
fat_supply_cols = ["Country", "Animal Products", "Cereals - Excluding Beer", "Eggs", "Fruits - Excluding Wine", \
             "Meat", "Milk - Excluding Butter", "Fish, Seafood", "Vegetables", "Obesity", "Undernourished", "Confirmed", "Deaths", "Recovered", "Population"]

# food_cols = ["Country", "Animal_Products", "Cereals", "Eggs", "Fruits", "Meat", "Milk", "Fish_Seafood", "Vegetables"]
fat_supply_transformed= fat_supply_df[fat_supply_cols].copy()

# Change column names
fat_supply_transformed = fat_supply_transformed.rename(columns={'Country': 'country', 
                                                    'Animal Products':'animal_products_fat',
                                                    'Cereals - Excluding Beer':'cereals_fat',
                                                    'Eggs':'eggs_fat',
                                                    'Fruits - Excluding Wine': 'fruits_fat',
                                                    'Meat':'meat_fat',
                                                    'Milk - Excluding Butter':'milk_fat',
                                                    'Fish, Seafood': 'fish_seafood_fat', 
                                                    'Vegetables':'vegetables_fat',
                                                    'Obesity': 'obesity',
                                                    'Undernourished': 'undernourished',
                                                    'Confirmed': 'confirmed_cases',
                                                    'Deaths': 'deaths',
                                                    'Recovered': 'recovered',
                                                    'Population': 'population'})
# Clean the data by dropping duplicates and setting the index
fat_supply_transformed.drop_duplicates("country", inplace=True)
fat_supply_transformed.set_index("country", inplace=True)

fat_supply_transformed
```

# Merge all of the Dataframes into one Dataframe

```python
dfs = [food_transformed, kcal_supply_transformed, fat_supply_transformed, protein_supply_transformed]
nan_value = 0

##EACH DATASET HAS THE SAME DATA IN OBESITY, UNDERNOURISHED, CONFIRMED_CASES, DEATHS, RECOVERED, AND POPULATION

covid_diet = functools.reduce(lambda left, right: pd.merge(left,right,on=['country', 'obesity', 'undernourished', 'confirmed_cases', 'deaths', 'recovered', 'population']), dfs)
covid_diet
```
